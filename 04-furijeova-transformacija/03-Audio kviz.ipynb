```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
from scipy import signal
from scipy.io import wavfile
from IPython import display
```

```python
audio_path = 'data/sine1.wav'
```

```python
display.Audio(audio_path)
```

```python
Fs, samples = wavfile.read(audio_path)
```

```python
Fs
```

```python
samples.shape
```

```python
f, t, s = signal.spectrogram(samples, Fs)
```

```python
plt.pcolormesh(t, f, np.log(s))
plt.colorbar()
plt.xlabel('Time [sec]')
plt.ylim(0, 2500)
plt.ylabel('Frequency [Hz]')
plt.show()
```
