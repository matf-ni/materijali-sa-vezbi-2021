# Matplotlib biblioteka

**Matplotlib** je Python biblioteka koja se koristi za 2D i 3D grafiku, kreiranje interaktivnih vizuelizacija i rad sa slikama. Na [ovom](https://matplotlib.org/) linku se nalazi zvanična stranica biblioteke, a [ovde](https://matplotlib.org/users/tutorials.html) možete pronaći koristan tutorijal.

Primeri koji slede imaju za cilj upoznavanje sa osnovnim funkcijama za kreiranje grafika, njihovo stilizovanje i anotaciju.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Primer 1. 

Potrebno je nacrtati grafik funkcije $y = sin(x)$ na ekvidistantnoj mreži od $100$ tačaka iz intervala od $[1, 10]$. 

Prvo ćemo formirati ekvidistantnu mrežu tačaka, a potom izračunati vrednosti funkcije.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

Podešavanjem svojstava `title`, `xlabel` i `ylabel` objekta `plt` vriši se postavljanje naslova i obeležja osa grafikona. Samo iscrtavanje se vrši funkcijom `plot` koja uz vrednosti parova tačaka `x` i `y` može imati i mnoga druga podešavanja. Funkcija `show` se koristi za prikaz grafika.

```python
# podesavanje naslova grafika i obelezja osa
plt.title('Grafik funkcije y=sin(x)')
plt.xlabel('x')
plt.ylabel('y')

# iscrtavanje grafika
plt.plot(x, y)

# prikaz grafika
plt.show()
```

### Primer 2.

Potrebno je nacrtati grafike funkcija $y = \sin (x)$ i $y = \cos (x)$ na intervalu $[-\pi, \pi]$. 

Prvo ćemo kreirati ekvidistantnu mrežu tačaka (npr. njih 100), a potom izračunati vrednosti funkcija. Vrednost konstante $\pi$ se može očitati iz `numpy` biblioteke.

```python
# ekvidistantna mreza sa vrednostima funkcija
N = 100
x = np.linspace(-np.pi, np.pi, N)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

Potom ćemo prikazati grafike trigonometrijskih funkcija korišćenjem pomenute funkcije `plot`. Ovoga puta iskoristićemo parametar `label` koji omogućava pridruživanje imena grafiku, a koji se kombinuje sa funkcijom `legend` koja se koristi za generisanje legende grafikona. Samu poziciju legende možemo kontrolisati parametrom `loc`. Uz imena, graficima trigonometrijskih funkcija pridružićemo i boje korišćenjem parametra `color`.  

```python
# iscrtavanja funkcija uz podesavanja boja prikaza i obelezja koja ce se javiti u legendi
plt.plot(x, y_sin, label='sin', color='blue')
plt.plot(x, y_cos, label='cos', color='green')

# postavljanje legende
plt.legend(loc='upper left')

# prikaz grafika
plt.show()
```

### Primer 3. 

U primeru koji sledi nacrtaćemo grafik funkcije $y = \sin(x)$ na intervalu $[-2\pi, 2\pi]$ sa podešavanjima stila linija i podeoka osa. 

Panel za iscrtavanje se može kontrolisati pomoću funkcije `figure`. Dimenzije panela se mogu eksplicitno zadati parametrom `figsize` sa širinom i visinom izraženom u inčima, a rezolucija rezultujućeg grafika parametrom `dpi` (engl. dots per inch). 

Opsezi osa se mogu zadati redom funkcijama `xlim` i `ylim`, a podeoci na njima funkcijama `xticks` i `yticks`. Videćemo u kasnijim primerima da se podeoci mogu i zarotirati.

Kontrola linija grafika se može postići preko parametara za podešavanje stila (parametar `linestyle`)  i debljine (parametar `linewidth`). 

```python
# podesavanje panela za iscrtavanje: podesavanje dimenzija (u incima) i rezolucije
plt.figure(figsize=(7, 5), dpi=80)

# postavljanje ogranicenja za podeoke na x i y osi
plt.xlim(-7, 7)
plt.ylim(-1, 1)

# postavljanje podeoka na x-osi i y-osi
plt.xticks(np.linspace(-7, 7, 15))
plt.yticks(np.linspace(-1, 1, 9))

# ekvidistantna mreza sa vrednostima funkcije
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.sin(x)

# iscrtavanje grafika sa kontrolom debljine i stila linije
plt.plot(x, y, linewidth='2', linestyle='--')

# cuvanje grafika
plt.savefig('sinus.jpg')

# prikaz grafika
plt.show()
```

U prethodnom primeru prikazano je i kako možemo sačuvati generisani grafik korišćenjem funkcije `savefig`. Ova funkcija ima mnogobrojne parametre za navođenje formata, rezolucije i slično.

### Primer 4.  Pitica.

Često je numeričke podatke potrebno prikazati i u formi pitica koje ukazuju na jasnu brojnost određenih grupa podataka. Sledeći primer ilustruje kako se može nacrtati pitica sa obeležjima. Pratićemo tri programska jezika i njihovu zastupljenost u jednoj anketi i koristićemo funkciju `pie` sa parametrima za podešavanje boje i labele.

```python
# zadavanje unosa
labels = ['JavaScript', 'Python', 'C#']
sizes = [50, 40, 15]
colors = ['green', 'yellow', 'orange']
```

```python
# podesavanje dimenzije grafika
plt.figure(figsize=(5, 5))

# podesavanje naslova grafika
plt.title('Zastupljenost programskih jezika')

# iscrtavanje pitice
plt.pie(sizes, colors=colors, labels=labels)

# prikaz pitice
plt.show()
```

Dodatni primeri se mogu pronaći na [ovoj](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.pie.html) adresi.

### Primer 5.  Toplotna mapa. 

Prikaz podataka u formi toplotne mape će značiti za vizuelizaciju trojki podataka (npr. prikaz jačine zvuka u zavisnosti od vremenske i frekvencijske komponente) ili za vizuelizaciju jačina korelacija između atributa.  

Za demonstraciju ćemo koristiti matricu dimenzija 10x10 sa slučajnim vrednostima iz uniformne raspodele [0, 1).

```python
A= np.random.rand(10, 10)
```

Toplotna mapa se kreira korišćenjem funkcije `imshow`. Boje koje se javljaju na mapi se kontrolišu parametrom `cmap`, a stubić sa intenzitetima koji su pridruženi nijansama boja se može dobiti pozivom funkcije `colorbar`. 

```python
# argumentom cmap se kontrolise shema boja npr. hot, ocean, ...
plt.imshow(A, cmap='viridis')

# prikaz stubica sa vrednostima
plt.colorbar()

# uklanjanje podeonih tacaka
plt.xticks([])
plt.yticks([])

# prikaz mape
plt.show()
```

Za više primera možete posetiti [ovaj](https://matplotlib.org/gallery/images_contours_and_fields/image_annotated_heatmap.html) link. 

### Primer 6. Stubići. 

Stubičaste grafikone ćemo često koristiti u radu sa kategoričkim vrednostima. Sledeći primer ilustruje prikaz imena osoba i njihovih godina. Za prikaz grafikona sa vertikalnim stubićima može se koristiti funkcija `bar`, dok se za prikaz grafikona sa horizontalnim stubićima može koristiti funkcija `barh`. 

```python
# liste kategorickih vrednosti
ages = [30, 20, 45]
names = ['ana', 'lazar', 'maja']

# generisanje stubica
plt.bar(names, ages, color='orange')

# postavljanje naslova
plt.title('Zastupljenost godina')

# prikaz stubica
plt.show()
```

### Primer 7.  Tačkasti grafikoni. 

Tačkaste grafikone često koristimo za uvid u zavisnosti među podacima. U primeru koji sledi kreiraćemo dva niza celih brojeva iz intervala [0, 20] i prikazati ih u vidu tačkastog grafikona uz podešavanje boja tačaka parametrom `color` i simbola za prikaz parametrom `marker`.

```python
# nizovi duzine 10 sa proizvoljnim elementima iz intervala [0, 20]
np.random.seed(7)
x = np.random.randint(low=0, high=20, size=10)
y = np.random.randint(low=0, high=20, size=10)

# generisanje tackastog grafikona
plt.scatter(x, y, color='green', marker='v')

# prikaz grafika
plt.show()
```
